```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
credits=pd.read_csv("tmdb_5000_credits.csv")
movies=pd.read_csv("tmdb_5000_movies.csv")
```

```python
credits.head()
```

```python
movies.head()
```

```python
print("credits:",credits.shape)
print("movies:",movies.shape)
```

```python
credits_rename=credits.rename(columns={"movie_id":"id"})
movies_merge=movies.merge(credits_rename,on='id')
movies_merge.head()
```

```python
movies_merge.drop(columns=['homepage','production_countries','title_x','title_y','status'],inplace=True)
movies_merge.shape
```

```python
movies_merge.isnull().sum()
```

# Using weighted Average

```python
R=movies_merge["vote_average"]
v=movies_merge['vote_count']
C=movies_merge["vote_average"].mean()
m=movies_merge['vote_count'].quantile(0.70)
```

```python
movies_merge['weighted_average']=((R*v)+(C*m))/(v+m)
movies_merge['weighted_average']
```

```python
movies_merge.head()
```

```python
a=movies_merge[['original_title','vote_average','vote_count','popularity','weighted_average']].sort_values('weighted_average',ascending=False).head(20)
a
```

```python

sns.barplot(x="weighted_average",y="original_title",data=a.sort_values('weighted_average',ascending=False).head(10))
plt.xlim(0, 9)

plt.title("Best Movies by average votes")
plt.xlabel('weigthed_average')
plt.ylabel('Movie titel')
plt.show()
```

```python
plt.figure(figsize=(12,6))
ax=sns.barplot(x="popularity",y="original_title",data=a.sort_values('popularity',ascending=False).head(10))
plt.title("Best Movies by popularity")
```

```python
from sklearn.preprocessing import MinMaxScaler
scaling=MinMaxScaler()
movies_merge['normalised_weighted_average']=scaling.fit_transform(movies_merge[['weighted_average']])
movies_merge['normalised_popularity']=scaling.fit_transform(movies_merge[['popularity']])
movies_merge.head()
```

```python
movies_merge['score']=movies_merge['normalised_weighted_average']*0.5+movies_merge['normalised_popularity']*0.5
b=movies_merge[['original_title','normalised_weighted_average','normalised_popularity','score']].sort_values('score',ascending=False).head(10)
b
```

```python
sns.barplot('score','original_title',data=b.sort_values('score',ascending=False).head(10))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfv=TfidfVectorizer(min_df=3, max_features=None, strip_accents='unicode',analyzer='word',token_pattern=r'\w{1,}',ngram_range=(1,3),stop_words='english')

movies_merge['overview']=movies_merge['overview'].fillna('')

tfv_matrix=tfv.fit_transform(movies_merge['overview'])
```

```python
print(tfv_matrix[:2])
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel

sig=sigmoid_kernel(tfv_matrix,tfv_matrix)
ind=pd.Series(movies_merge.index,index=movies_merge['original_title']).drop_duplicates()
sig[:1]
```

```python
def recommendation(title,sig=sig):
    indx=ind[title]
    sig_score=list(enumerate(sig[indx]))
    
    sig_score=sorted(sig_score,key=lambda x:x[1],reverse=True)
    
    sig_score=sig_score[1:11]
    
    Movie_index=[i[0] for i in sig_score]
   
    return movies_merge['original_title'].iloc[Movie_index]

```

```python
recommendation('Titanic')
```
